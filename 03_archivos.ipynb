```python
#crear un archivo y escribir una contraaseña dentro

def manejar_archivo(nombre_archivo, modo, contenido=None):
    if modo not in ['r', 'w', 'a', 'x']:
        raise IOError("Modo no soportado. Usa 'r', 'w', 'a' o 'x'.")
    try:
        with open(nombre_archivo, modo) as archivo:
            if modo == 'r':
                return archivo.read()
            else:
                archivo.write(contenido)
                return f"Contenido escrito en {nombre_archivo} en modo {modo}"
    except FileNotFoundError:
        return f"El archivo {nombre_archivo} no existe y no puede ser leído."
   
print(manejar_archivo("contraseña.txt", "w", "P@sw0rd123"))
print(manejar_archivo("contraseña.txt", "r"))
 
```

```python
import random

class Carta:
    def __init__(self, valor, figura):
        self.valor = valor
        self.figura = figura

    def __repr__(self):
        return f"{self.valor} de {self.figura}"

class Baraja:
    figuras = ['Corazones', 'Diamantes', 'Tréboles', 'Picas']
    valores = ['As', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jota', 'Reina', 'Rey']

    def __init__(self):
        self.cartas = [Carta(valor, figura) for figura in self.figuras for valor in self.valores]

    def barajar(self):
        random.shuffle(self.cartas)

    def mostrar(self):
        for carta in self.cartas:
            print(carta)

# Baraja original
baraja = Baraja()
print("Baraja original:\n")
baraja.mostrar()

# Barajada
baraja.barajar()
print("\nBaraja barajado:")
baraja.mostrar()
 
 
```
