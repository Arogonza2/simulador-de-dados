Escribe un programa que permita generar la tabla de multiplicar de un numero entero positivo N, comenzando desde 1.
Si el usuario escribe un numero incorrecto, el programa no se ejecuta. 
En cambio pregunta de nuevo por l ainformacion hasta que el dato ingresado sea correcto  

```python
comprobar = True

while comprobar == True:

    n = int(input("Indique un numero entero positivo:"))

    if n > 0 :
                
        for i in range(1,11):
            
            print(n,"por",i ,"es igual a:", n*i)
            
        comprobar = False
        
    else:
        print(" El numero introducido no es correcto, intentelo de nuevo. ")
```

Escribe un programa que, al recibir como dato un numero entero positivo N calcule el resultado de la siguiente serie:

1 + (1/2) + (1/3) + (1/5), + ..... + (1/N)

Si el usuario escribe un numero incorrecto, el programa no se ejecuta. 
En cambio pregunta de nuevo por la informacion hasta que el dato ingresado sea correcto 

```python

comprobar = True

while comprobar == True:

    n = int(input("Indique un numero entero positivo:"))

    if n > 0 :
        resultado = 0
        
        for i  in range(1,n+1):
            resultado += (1/i)
        print(" El resultado de la serie es: ", resultado)
         
        comprobar = False   
            
    else:
            print(" El numero introducido no es correcto, intentelo de nuevo. ")
```

Escribe un programa que, al recibir como dato un numero entero positivo N calcule el resultado de la siguiente serie:

1 / (1/2) * (1/3) / (1/4), * ..... (*/) (1/N)

Si el usuario escribe un numero incorrecto, el programa no se ejecuta. 
En cambio pregunta de nuevo por la informacion hasta que el dato ingresado sea correcto 

```python
comprobar = True

while comprobar == True:

    n = int(input("Indique un numero entero positivo:"))

    if n > 0 :
        resultado = 1
        
        for i  in range(2,n+1):
            
            if i % 2 == 0:
                resultado= resultado / (1/i)
            else:
                resultado = resultado * (1/i)
                   
        print(" El resultado de la serie es: ", resultado)
         
        comprobar = False   
            
    else:
            print(" El numero introducido no es correcto, intentelo de nuevo. ")
```

Crea un programa que, al recibir como dato un numero entero N, obtenga el resultado de la siguiente serie:

1 - 2 + 3 - .... +- N^N 

Si el usuario escribe un numero incorrecto, el programa no se ejecuta. 
En cambio pregunta de nuevo por la informacion hasta que el dato ingresado sea correcto 

```python


resultado = 0
comprobar = True

while comprobar == True:
     
     n = int(input(" Introduce un numero de terminos de la serie: "))

     if n > 0:
          
          comprobar = False
          
          for i in range(1, n+1):
               
               if i % 2 == 0:
                    resultado -= i**i
               else:
                    resultado += i**i
                    
          print(" El resultado de la serie es : ", resultado)
     

     else:
          print(" El numero ingresado no es correcto, intentalo de nuevo.")
```
